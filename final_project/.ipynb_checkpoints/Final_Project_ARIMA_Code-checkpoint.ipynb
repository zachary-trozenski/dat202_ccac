```python
"""
This is my ARIMA code to forecast rainfall in Pittsburgh.
The data has been acquired via NOAA from 2004 to 2014. It has been recorded daily and hourly
within each given day. There are two recording stations given in the data, Murrysville and Pittsburgh.
Further inquiry will be needed to understand the change from former to the latter.

Data Dictionary:
Source : https://www1.ncdc.noaa.gov/pub/data/cdo/documentation/PRECIP_HLY_documentation.pdf
Station - Station ID number
Station - Station name (string)
Date - YYYYMMDD HH:MM
HPCP - Hourly percipitation total (1/100 inches)

References:
Brownlee - Introduction to Time Series Forecasting with Python
Stack Overflow

I had initially built this model with very basic alterations to my data (normalizing dates and fixing NaN values)
however once I had fit my model and tried to predict future values, it became clear that the dataset as it was
composed was not going to be able to be used for predictions.
To be able and rectify errors (specifically: ValueWarning: A date index has been provided, but it is not 
monotonic and so will be ignored when e.g. forecasting.' forecasting.', ValueWarning)) I did some preprocessing
with the goal of remediating the error and trying to get predictions. For more details see `proc_log.md`
"""
```

```python
# Import the test data
test_rain = pd.read_csv('./post_processed_data/daily_rain_13.csv', parse_dates=['DATE'])
```

```python
# post-processing to join the forecast to the dates in the test set and create a dataframe
test_dates = test_rain['DATE']
rain_forecast = pd.DataFrame({"DATE" : test_dates, "HPCP" : forecast.values})
rain_forecast.head()
```

```python
## Let's plot the prediction versus the actual side by side and see what happens!

graphs = pyplot.figure()
graphs.set_figheight(7)
graphs.set_figwidth(21)

ax0 = graphs.add_subplot(121)
ax0.plot_date(x=test_rain["DATE"], y=test_rain["HPCP"], fmt="b-")
ax0.plot_date(x=rain_forecast["DATE"], y=rain_forecast["HPCP"], fmt="r-")
ax0.xaxis.set_tick_params(rotation=30, labelsize=10)

ax1 = graphs.add_subplot(122)
ax1.plot_date(x=rain_forecast["DATE"], y=rain_forecast["HPCP"], fmt="r-")
ax1.xaxis.set_tick_params(rotation=30, labelsize=10)

pyplot.show()
```

```python
# Whoa, something might be wrong here. According to several helpful internet sources this isn't really bad.
# The data doesn't exhibit any specific trends so a static mean observation is actually a best guess for
# how much it will rain on any given future day.
# Let's calculate the mean squared errore and the mean absolute error.
mse = mean_squared_error(y_true=test_rain['HPCP'], y_pred=rain_forecast['HPCP'])
mae = mean_absolute_error(y_true=test_rain['HPCP'], y_pred=rain_forecast['HPCP'])

print(mse ,'\n', mae)

# These look pretty good, but can we do better? Can we actually try to predict differing rain values if we include
# seaonality?
```

```python
# Let's check first to see if our data adheres to the properties of stationarity.
# We can test this by running the augemented dickey fuller test (tests whether or not data is stationary)
# We want to get the p-value to see whether or not the null hypothesis can be rejected (null hypothesis here
# is that the test set is not stationary)

from statsmodels.tsa.stattools import adfuller

print("p-value:", adfuller(rain_data)[1])
```

```python
# The p-value indicates that the null hypothesis can be rejected therefore meaning our data is stationary
# No need to difference the data (stabilizing the data, mean and variance)
```

```python
## Now that we know differencing isn't needed another thing we want to do is find the optimal order for our 
# ARIMA model. Where earlier the order was set based on some loose research and a guess, 
# this next process will conduct a test on the different order parameters to determine the best fit. I'm 
# including seasonality and the period of differencing (m) here.

import pmdarima as pmd
#from statsmodels.compat.python import iterkeys

auto_model = pmd.auto_arima(rain_data, start_p=1, start_q=1, test="adf", trace=True, seasonal=True, m=7)

# Setting an m value to 365 causes a crash. Might be that my machine doesn't have the compute power or
# the software is not designed for this. Instead I've elected to set m to 7 which in the documentation refers
# to daily. I'll continue tinkering with this to see what happens.
```

```python
## The best model from the auto_arima function returns (1,0,1)(1,0,1,7) (m=7)with that noted I'm going to train 
# a new model with the given order.

## Training and fitting the model
b_model = ARIMA(rain_data, order=(1, 0, 1), seasonal_order=(1,0,1,7), freq='D')
b_fit_model = b_model.fit()
```

```python
print(b_fit_model.summary())
```

```python
# Assigned the residual errors to a variable and plot them
res2 = pd.DataFrame(b_fit_model.resid)
res2.plot()
pyplot.show()
```

```python
# Check the distribution of the residual errors
res2.plot(kind='kde')
pyplot.show()
```

```python
print(res2.describe())
# This new model does have a bit more of a bias to it than before as evidenced by the mean value
```
